```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Product_A_Sales': [10, 20, 15, 25, 30],
    'Product_B_Sales': [5, 15, 10, 20, 25],
    'Product_C_Sales': [8, 18, 12, 22, 28]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Line graph
plt.figure(figsize=(10, 6))
for product in ['Product_A_Sales', 'Product_B_Sales', 'Product_C_Sales']:
    plt.plot(df['Month'], df[product], marker='o', label=product)
plt.title('Monthly Sales of Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# Bar graph
plt.figure(figsize=(10, 6))
bar_width = 0.25
months = df['Month']
bar1 = df['Product_A_Sales']
bar2 = df['Product_B_Sales']
bar3 = df['Product_C_Sales']

r1 = range(len(months))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, bar1, color='blue', width=bar_width, edgecolor='grey', label='Product A')
plt.bar(r2, bar2, color='green', width=bar_width, edgecolor='grey', label='Product B')
plt.bar(r3, bar3, color='red', width=bar_width, edgecolor='grey', label='Product C')

plt.xlabel('Month', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(months))], months)
plt.ylabel('Sales')
plt.title('Sales Comparison of Products')
plt.legend()
plt.show()

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Product_A_Sales'], df['Product_B_Sales'], color='blue', label='A vs B')
plt.scatter(df['Product_A_Sales'], df['Product_C_Sales'], color='green', label='A vs C')
plt.scatter(df['Product_B_Sales'], df['Product_C_Sales'], color='red', label='B vs C')
plt.title('Scatter Plot of Product Sales')
plt.xlabel('Sales of Product A / B')
plt.ylabel('Sales of Product B / C')
plt.legend()
plt.grid(True)
plt.show()
```
